```python
import pandas as pd
import numpy as np
prices = pd.read_csv("./data/sample_prices.csv")
returns= prices.pct_change()
returns
```

```python
returns = returns.dropna()
returns
```

```python
returns.std()
```

```python
devations = returns - returns.mean()
squared_devations = pow(devations, 2)
variance = squared_devations.mean()
volatility = np.sqrt(variance)
```
